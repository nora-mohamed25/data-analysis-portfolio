Customer_Purchase

Predict the likelihood of a customer purchasing a product using the "KNN Classification" model.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
import pandas as pd
df = pd.read_csv('/content/customer_purchase_data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df[['Age', 'AnnualIncome', 'NumberOfPurchases', 'TimeSpentOnWebsite', 'DiscountsAvailed', 'PurchaseStatus']]
df.head()
```

```python
x = df.drop('PurchaseStatus', axis=1)
y = df['PurchaseStatus']
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)
```

```python
y_pred = knn_model.predict(x_test)
```

```python
y_pred
```

```python
comparison_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
comparison_df
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("The Accuracy Is:", accuracy)
```
