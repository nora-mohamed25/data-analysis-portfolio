```python
import re
import string
from nltk.corpus import stopwords
from nltk import word_tokenize , pos_tag
from sklearn.feature_extraction.text import TfidfVectorizer
import nltk
```

```python
nltk.download('all')
```

```python
text_input = "I am a motivated, and proactive individual with strong leadership skills!, and a passion for continuous learning.2026"
```

```python
tokens = word_tokenize(text_input)
tokens
```

```python
tokens = [token for token in tokens if token not in string.punctuation]
tokens
```

```python
stop_words = set(stopwords.words('english'))
tokens = [token for token in tokens if token not in stop_words]
tokens
```

```python
tokens = [re.sub('[^a-zA-Z]','',token) for token in tokens if re.sub('[^a-zA-Z]','',token)]
tokens
```

```python
tokens = [token for token in tokens if len(token) > 1]
tokens
```

```python
procoessed = ' '.join(tokens)
procoessed
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
for token in tokens:
  stemmed = stemmer.stem(token)
  print(stemmed)
```
