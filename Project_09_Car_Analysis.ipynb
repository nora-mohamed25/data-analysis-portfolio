#CARS

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('measurements.csv')

```

```python
df.head()
```

```python
df.info()
```

```python

df["distance"] = df["distance"].str.replace(",", ".").astype(float)
```

```python

df["consume"] = df["consume"].str.replace(",", ".").astype(float)
```

```python

df["temp_inside"] = df["temp_inside"].str.replace(",", ".").astype(float)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop("specials", axis=1, inplace=True)
df.drop("refill liters", axis=1, inplace=True)
df.drop("refill gas", axis=1, inplace=True)
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["temp_inside"].fillna(df["temp_inside"].mean(), inplace=True)
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=["gas_type"])
```

```python
df.head()
```

```python
df.info()
```

```python
x = df.drop("consume", axis=1)
y = df["consume"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
```

```python
y_pred = linear_model.predict(x_test)
y_pred
```

```python
df.compare = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
df.compare
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```
